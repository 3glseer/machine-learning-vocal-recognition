```python
import numpy as np

PATH = '../data/train/npy/'
FILE_NAME = 'train'
N = 39
LABELS_END_WITH = '.labels.npz'
DATA_END_WITH = '.npy'
```

```python
with open(PATH + 'feature_names.txt', 'r') as f:
    feature_names = f.read().splitlines()
```

### Feature format
+ 270 {30 MFCCs + 30∆ + 30∆∆} (90 means, 90 variances, 90 ranges)
+ 51 spectral contraction (17 means, 17 variances, 17 ranges)
+ 51 spectral flatness (17 means, 17 variances, 17 ranges)
+ 51 fluctogram (17 means, 17 variances, 17 ranges)
+ 51 spectral energy (17 means, 17 variances, 17 ranges)
+ 51 spectral power (17 means, 17 variances, 17 ranges)
+ 51 spectral flux (17 means, 17 variances, 17 ranges)
+ 51 spectral centroid (17 means, 17 variances, 17 ranges)

```python
data_files = [PATH + FILE_NAME + format(i, '02d') + DATA_END_WITH for i in range(1, N+1)]
label_files = [PATH + FILE_NAME + format(i, '02d') + LABELS_END_WITH for i in range(1, N+1)]
```

```python
npylist = [np.load(e, mmap_mode='r') for e in data_files]
data = np.concatenate(npylist)
data.shape
```

```python
npylist = [np.load(e, mmap_mode='r') for e in label_files]
female_labels = np.concatenate([e['female'][:,0] for e in npylist])
male_labels = np.concatenate([e['male'][:,0] for e in npylist])
choral_labels = np.concatenate([e['choral'][:,0] for e in npylist])
labels = (female_labels, male_labels, choral_labels)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, female_labels, train_size = 0.1, test_size=0.05, random_state=42)
```

```python
clf = svm.SVR(kernel ="linear")
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
def tipple_heatmap(data, start, length, title):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,7))
    test = data.iloc[:,start:start+length]
    corr = test.corr()
    fig.suptitle(title, fontsize=16)

    ax1.set_title("mean")

    ht1 = sns.heatmap(
        corr, 
        vmin=-1, vmax=1, center=0,
        cmap=sns.diverging_palette(20, 220, n=200),
        square=True,
        annot=False,
        ax = ax1
    )
    ht1.set_xticklabels(
        ax.get_xticklabels(),
        rotation=90,
        horizontalalignment='right'
    );

    test = data.iloc[:,start+length:start+2*length]

    corr = test.corr()

    ax2.set_title("var")

    ht2 = sns.heatmap(
        corr, 
        vmin=-1, vmax=1, center=0,
        cmap=sns.diverging_palette(20, 220, n=200),
        square=True,
        annot=False,
        ax = ax2
    )
    ht2.set_xticklabels(
        ax.get_xticklabels(),
        rotation=90,
        horizontalalignment='right'
    );

    test = data.iloc[:,start+2*length:start+3*length]

    corr = test.corr()
    ax3.set_title("range")
    ht3 = sns.heatmap(
        corr, 
        vmin=-1, vmax=1, center=0,
        cmap=sns.diverging_palette(20, 220, n=200),
        square=True,
        annot=False,
        ax = ax3
    )
    ht3.set_xticklabels(
        ax.get_xticklabels(),
        rotation=90,
        horizontalalignment='right'
    );
    fig.tight_layout()
```

```python
tipple_heatmap(data, 270, 17, 'spectral contraction')
tipple_heatmap(data, 321, 17, 'spectral flatness')
tipple_heatmap(data, 372, 17, 'fluctogram')
tipple_heatmap(data, 423, 17, 'spectral energy')
tipple_heatmap(data, 474, 17, 'spectral power')
tipple_heatmap(data, 525, 17, 'spectral flux')
tipple_heatmap(data, 576, 17, 'spectral centroid')
```
